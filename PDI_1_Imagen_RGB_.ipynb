```python

                   # PRACTICO 1. PDI: IMÁGENES RGB



#  1. IMPORTO LAS LIBRERIAS
import numpy as np                      #libreria para operecaiones numéricas
import matplotlib.pyplot as plt         #libreria para graficar
import imageio                          #libreria para acceder a una imagen como matriz


# 2.LEO LA IMAGEN
imagen= imageio.imread('/content/drive/MyDrive/IMAGENES- PDI_1/zoologico.jpg')  


# 3. MUESTRO LA IMAGEN EN FORMA MATRICIAL
print ("")
print("REPRESENTACIÓN DE LA IMAGEN EN FORMA MATRICIAL")
print ("")
print (imagen)

# 4. MUESTRO LA IMAGEN EN PANTALLA
print("2- NOMBRE DE LA IMAGEN A MODIFICAR: Zoológico")
plt.imshow(imagen)

# 3. DIMENSIONES DE LA IMAGEN ORIGINAL
print ("")
print("3- DIMENSIONES DE LA IMAGEN")
print("(vertical, horizontal, Nro de canales)")
print(imagen.shape)
```

```python
# 4. EXTRAER LOS 3 CANALES DE LA IMAGEN (RGB)

print("4- EXTRACCIÓN DE LOS 3 CANALES (RGB)")

R=imagen[:,:,0]
G=imagen[:,:,1]
B=imagen[:,:,2]
print("")
```

```python
# SUBPLOTS PARA VER LOS TRES CANALES (RGB) EN LA MISMA IMAGEN 

fig, ax = plt.subplots(1, 3, dpi=300)
ax[0].imshow(R, cmap='gray')
ax[0].set_axis_off()
print('canal R, CANAL G, CANAL B')
ax[1].imshow(G, cmap='gray')
ax[1].set_axis_off()
ax[2].imshow(B, cmap='gray')
ax[2].set_axis_off()
plt.show()
```

```python
# 5 - MUESTRO EL CANAL R
print("CANAL R")
plt.imshow(R, cmap='gray')
print("dimension del canal R:", R.shape)

# 5.1- MODIFICO EL CANAL R con scroll

import ipywidgets as widgets
slider = widgets.IntSlider(value=8, max=500)
display(slider)

R= R.astype(np.float)            #declaro R como un flotante, para que reciba cualquier resultado, incluso hasta decimales
modr=R*slider.value              #EN LA VARIABLE 'modr', almaceno el valor del producto entre el canal R y el valor que se guarda en en slider
modr=np.clip(modr,0,255)         #restringo  que el valor  modr este en Unit8 para q el valor vaya de o a 255 y no distrorcione la imagen
plt.imshow(modr, cmap='gray')    #grafico la imagen que se modificó al multipkicar el valor del scroll con R
```

```python
# 6 - MUESTRO EL CANAL G
print("CANAL G")
plt.imshow(G, cmap='gray')
print("dimension del canal G:", G.shape)


#6.1- MODIFICO EL CANAL G con scroll

import ipywidgets as widgets
slider = widgets.IntSlider(value=10, max=500)
display(slider)

G= G.astype(np.float)             #declaro G como un flotante, para que reciba cualquier resultado, incluso hasta decimales
modg=G*slider.value               #EN LA VARIABLE 'modg', almaceno el valor del producto entre el canal G y el valor que se guarda en en slider
modg=np.clip(modg,0,255)          #restringo  que el valor  modg este en Unit8 para q el valor vaya de o a 255 y no distrorsione la imagen
plt.imshow(modg, cmap='gray')     #grafico la imagen que se modificó al multiplicar el valor del scroll con G
```

```python
# 7 - MUESTRO EL CANAL B
print("CANAL B")
plt.imshow(B, cmap='gray')
print("dimension del canal B:", B.shape)

#7.1- MODIFICO EL CANAL B con scroll

import ipywidgets as widgets
slider = widgets.IntSlider(value=5, max=500)
display(slider)

B= B.astype(np.float)           #declaro B como un flotante, para que reciba cualquier resultado, incluso hasta decimales
modb=B*slider.value             #EN LA VARIABLE 'modb', almaceno el valor del producto entre el canal B y el valor que se guarda en en slider
modb=np.clip(modb,0,255)        #restringo  que el valor  modb este en Unit8 para q el valor vaya de o a 255 y no distrorsione la imagen
plt.imshow(modb, cmap='gray')   #grafico la imagen que se modificó al multiplicar el valor del scroll con B
```

```python
# 8. GUARDO LA IMAGEN 

nuevaimagen=np.zeros((570, 1000, 3))    #genero una nueva matriz vacia, con las dimensiones de la imagen original, para mostrar la imagen modificada en RGB 

nuevaimagen[:,:,0]=modr[:,:]            # En cada uno de los canales RGB, guardo las modificaciones anteriores
nuevaimagen[:,:,1]=modg[:,:]
nuevaimagen[:,:,2]=modb[:,:]

imageio.imsave('/content/drive/MyDrive/IMAGENES- PDI_1/zoologico4.jpg', nuevaimagen)    #guardo la imagen

# MUESTRO LA IAMGEN EN LOS TRES CANALES

nuevaimagen=nuevaimagen.astype(np.uint8)
plt.imshow(nuevaimagen)
```
